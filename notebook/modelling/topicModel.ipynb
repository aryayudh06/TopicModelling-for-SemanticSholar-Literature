```python
import pandas as pd

df = pd.read_csv("../../data/processed_semantic_data.csv")
```

```python
df = df.head(1)
df
```

```python
df.to_csv("head_semantic_results.csv")
```

```python
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer
from sentence_transformers import SentenceTransformer

# Misal kolom teks bernama 'content'
texts = df["content"].dropna().tolist()

# Buat model embedding
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# Inisialisasi BERTopic
topic_model = BERTopic(embedding_model=embedding_model)

# Training model
topics, probs = topic_model.fit_transform(texts)
```

```python
coherence = topic_model.get_coherence()
print(f"Coherence Score: {coherence}")
```

```python
# Simpan model ke file
topic_model.save("./model/")
```
